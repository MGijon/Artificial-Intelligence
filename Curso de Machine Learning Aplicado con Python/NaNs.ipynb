```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load the data
data = pd.read_csv('Data/clean.csv')
data.head(2)
```

```python
data.notnull().apply(pd.Series.value_counts)
```

```python
(data != 0).apply(pd.Series.value_counts)
```

```python
available = ( (data != 0) & (data.notnull()) )
available.head()
```

```python
#available.all(axis=1) # ¿qué filas están completas?
available.all(axis=1).value_counts()
```

```python
# perderiamos demasiados datos, vamos a volver a un punto anterior
mask = available['worldwide_gross']
data = data[mask]
```

```python
( (data != 0) & (data.notnull()) ).worldwide_gross.value_counts()
```

```python
from sklearn.preprocessing import Imputer
```

```python
data.head(2) # observams que tenemos que sacar las variables de texto y que al hacer el merge se nos ha
             # duplicado una columna, las eliminaremos
```

```python
data = data.drop('movie_title', axis = 1)
data = data.drop('duration.1', axis = 1)
```

```python
data.head()
```

```python
imputer = Imputer(missing_values = np.nan,
                  strategy = 'mean',
                  axis = 1) # remplazará los valores por las medias de su columna
values = imputer.fit_transform(data)
```

```python
data.values
```

```python
values
```

```python
X = pd.DataFrame(values)
X.columns = data.columns
X.index = data.index
```

```python
X.head(4)
```

```python
X.to_csv('Data/intermediate_resutls.csv')
```
