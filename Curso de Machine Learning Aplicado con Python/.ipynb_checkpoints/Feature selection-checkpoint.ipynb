```python
import numpy as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# loading data
X = pd.read_csv('Data/intermediate_resutls.csv')
X.head(2)
```

## Visualizar interdependencia entre las variables

```python
y = X['worldwide_gross']
X= X.drop('worldwide_gross', axis = 1)
X.head(2)
```

```python
X.columns[0]
```

```python
X = X.drop(X.columns[0], axis = 1)
X.head(2)
```

```python
sns.heatmap(X.corr())
```

## Correlación con la variable respuesta

```python
Z = pd.concat([X, y], axis = 1)
sns.pairplot(Z)
```

```python
clase = pd.cut(X['production_budget'], 8).cut.codes.rename('class')
Z2 = pd.concat([X, clase], axis=1).head()
```

```python
sns.pairplot(Z2, hue='class') # pinta según los valores de clase
```
